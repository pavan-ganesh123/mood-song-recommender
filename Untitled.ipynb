```python
import os
import pandas as pd
import joblib

# Load the trained model
model = joblib.load("mood_classifier.pkl")

# Load songs dataset
songs_df = pd.read_csv("songs.csv")

# Function to play a song using CMD
def play_song(file_path):
    if not os.path.exists(file_path):
        print(f"Error: File not found - {file_path}")
        return

    print(f"Playing: {file_path}")
    os.system(f'start "" "{file_path}"')  # Open file using default media player

# Function to predict mood and play a song
def predict_mood_and_play(message):
    predicted_mood = model.predict([message])[0]
    mood_name = {0: "sadness", 1: "joy", 2: "love", 3: "anger", 4: "fear", 5: "surprise"}[predicted_mood]

    # Fetch songs matching the predicted mood
    filtered_songs = songs_df[songs_df["moods"].str.contains(mood_name, case=False, na=False)]
    
    if not filtered_songs.empty:
        recommended_songs = filtered_songs.sample(min(3, len(filtered_songs)))  # Pick up to 3 songs
    else:
        recommended_songs = songs_df.sample(3)  # Fallback: Pick any 3 random songs

    for _, row in recommended_songs.iterrows():
        song_name = row["song_name"]
        file_path = row["file_path"].strip()  # Remove extra spaces
        file_path = file_path.replace("//", "/").replace("\\", "/")  # Normalize slashes
        file_path = os.path.abspath(file_path)

        if not os.path.exists(file_path):
            print(f"Error: File not found - {file_path}")
            continue  # Skip this song

        print(f"- Playing: {song_name} (Path: {file_path})")
        play_song(file_path)  # Play the song

# Test the function
predict_mood_and_play("I am so tired of things")
```

```python
import os
import pandas as pd
import joblib
import pygame
import time
import vlc
# Load the trained model
model = joblib.load("mood_classifier.pkl")

# Load songs dataset
songs_df = pd.read_csv("songs.csv")

# Initialize pygame mixer
pygame.mixer.init()

def play_song(file_path):
    """Play a song using VLC for better compatibility."""
    if not os.path.exists(file_path):
        print(f"Error: File not found - {file_path}")
        return False

    player = vlc.MediaPlayer(file_path)
    player.play()

    print(f"Now Playing: {file_path}")
    print("Controls: [P]ause | [R]esume | [N]ext (Skip) | [Q]uit")

    while True:
        command = input("Enter command: ").strip().lower()
        if command == "p":
            player.pause()
            print("Paused")
        elif command == "r":
            player.play()
            print("Resumed")
        elif command == "n":
            player.stop()
            print("Skipped")
            return True  # Skip to next song
        elif command == "q":
            player.stop()
            print("Stopped playing.")
            return False  # Quit playback

    return True  # Proceed to next song

def predict_mood_and_play(message):
    """Predict mood from message and play all matching songs."""
    predicted_mood = model.predict([message])[0]
    mood_name = {0: "sadness", 1: "joy", 2: "love", 3: "anger", 4: "fear", 5: "surprise"}[predicted_mood]

    # Fetch songs matching the predicted mood
    filtered_songs = songs_df[songs_df["moods"].str.contains(mood_name, case=False, na=False)]
    
    if filtered_songs.empty:
        print(f"No songs found for mood: {mood_name}")
        return

    print(f"Playing {len(filtered_songs)} songs for mood: {mood_name}")
    
    for _, row in filtered_songs.iterrows():
        file_path = os.path.abspath(row["file_path"].strip())
        if not os.path.exists(file_path):
            print(f"Error: File not found - {file_path}")
            continue  # Skip missing files
        
        if not play_song(file_path):  # If the user chooses to quit
            break

# Test the function
predict_mood_and_play("I don't quit")
```

```python
import pandas as pd
import nltk
import joblib
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Load dataset (Modify nrows based on your dataset size)
df = pd.read_csv("mood_data.csv", encoding="utf-8", nrows=416800)

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Text Preprocessing Function
def preprocess_text(text):
    tokens = word_tokenize(text.lower())  # Convert to lowercase and tokenize
    tokens = [word for word in tokens if word.isalpha()]  # Remove punctuation/numbers
    tokens = [word for word in tokens if word not in stopwords.words('english')]  # Remove stopwords
    tokens = [lemmatizer.lemmatize(word) for word in tokens]  # Apply lemmatization
    return " ".join(tokens)

# Apply preprocessing
df["text"] = df["text"].apply(preprocess_text)

# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(df["text"], df["label"], test_size=0.2, random_state=42)

# Define different models
models = {
    "Naive Bayes": MultinomialNB(),
    "Logistic Regression": LogisticRegression(max_iter=500),
    "SVM": SVC(kernel="linear", probability=True)
}

# TF-IDF Vectorizer with bigrams
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2))

# Transform the data
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Train and evaluate models
best_model = None
best_accuracy = 0
best_model_name = ""

for name, model in models.items():
    model.fit(X_train_tfidf, y_train)
    y_pred = model.predict(X_test_tfidf)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{name} Accuracy: {accuracy * 100:.2f}%")

    # Save the best model
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = model
        best_model_name = name

print(f"\nBest Model: {best_model_name} with {best_accuracy * 100:.2f}% accuracy")

# Save the best model
joblib.dump(best_model, "best_mood_classifier.pkl")
joblib.dump(vectorizer, "tfidf_vectorizer.pkl")

print("\nBest model saved as 'best_mood_classifier.pkl'")
print("TF-IDF vectorizer saved as 'tfidf_vectorizer.pkl'")
```

```python
import pandas as pd
import nltk
import joblib
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import StackingClassifier
from sklearn.metrics import accuracy_score

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Load dataset
df = pd.read_csv("mood_data.csv", encoding="utf-8", nrows=416800)

# Filter dataset to include only labels 0, 1, 2, 3 (Sad, Happy, Love, Anger)
df = df[df["label"].isin([0, 1, 2, 3])]

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Text Preprocessing Function
def preprocess_text(text):
    tokens = word_tokenize(str(text).lower())  # Convert to lowercase and tokenize
    tokens = [word for word in tokens if word.isalpha()]  # Remove punctuation/numbers
    tokens = [word for word in tokens if word not in stopwords.words('english')]  # Remove stopwords
    tokens = [lemmatizer.lemmatize(word) for word in tokens]  # Apply lemmatization
    return " ".join(tokens)

# Apply preprocessing
df["text"] = df["text"].apply(preprocess_text)

# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(df["text"], df["label"], test_size=0.2, random_state=42)

# TF-IDF Vectorizer with bigrams
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Define base models
base_models = [
    ("Naive Bayes", MultinomialNB()),
    ("Logistic Regression", LogisticRegression(max_iter=500)),
    ("SVM", SVC(kernel="linear", probability=True))
]

# Define stacking classifier
stacking_model = StackingClassifier(estimators=base_models, final_estimator=LogisticRegression())

# Train model
stacking_model.fit(X_train_tfidf, y_train)
y_pred = stacking_model.predict(X_test_tfidf)
accuracy = accuracy_score(y_test, y_pred)
print(f"Stacking Model Accuracy: {accuracy * 100:.2f}%")

# Save the model and vectorizer
joblib.dump(stacking_model, "best_mood_classifier.pkl")
joblib.dump(vectorizer, "tfidf_vectorizer.pkl")

print("\nBest model saved as 'best_mood_classifier.pkl'")
print("TF-IDF vectorizer saved as 'tfidf_vectorizer.pkl'")

# Predict function for multiple sentences
def predict_mood(text):
    sentences = sent_tokenize(text)  # Split text into sentences
    sentence_vectors = vectorizer.transform([preprocess_text(sentence) for sentence in sentences])
    predictions = stacking_model.predict(sentence_vectors)
    mood = np.bincount(predictions).argmax()  # Get the most frequent mood
    mood_labels = {0: "sad", 1: "happy", 2: "love", 3: "anger"}
    return mood_labels[mood]

# Test with multiple sentences
print("\nPredicted Mood:", predict_mood("I feel so down today. Nothing is going well."))
```
